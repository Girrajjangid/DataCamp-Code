```python
# Convert the Country column to a one hot encoded Data Frame
one_hot_encoded = pd.get_dummies(raw_data, columns=['Country'], prefix='OH')

# Print the columns names
print(one_hot_encoded.columns.tolist())
```

```python
# Create a series out of the Country column
countries = raw_data.Country

# Get the counts of each category
country_counts = countries.value_counts()

# Print the count values for each category
print(country_counts)
```

```python
# Create a mask for only categories that occur less than 10 times
mask = countries.isin(country_counts[country_counts < 300].index) # countries name which is less than 10 entry

# Print the top 5 rows in the mask series
print(mask)
```

```python
# Label all other categories as Other
countries[mask] = 'Other'

# Print the updated category counts
print(countries.value_counts())
```

# Binning Numerical data

```python
# cutting data into groups
raw_data.Age.value_counts()
```

```python
raw_data.Age.head()
```

```python
pd.cut(raw_data.Age , bins=[-np.inf , 0 ,2,np.inf] , labels = [1,2,3])
```

```python
# Create the Paid_Job column filled with zeros
raw_data['Paid_Job'] = 0

# Replace all the Paid_Job values where ConvertedSalary is > 0
raw_data.loc[raw_data['ConvertedSalary'] > 0 ,'Paid_Job' ] = 1
# Print the first five rows of the columns
print(raw_data[['Paid_Job', 'ConvertedSalary']].head())
```

```python
    # Bin the continuous variable ConvertedSalary into 5 bins
raw_data['equal_binned'] = pd.cut(raw_data['ConvertedSalary'],bins=5)

# Print the first 5 rows of the equal_binned column
print(raw_data[['equal_binned', 'ConvertedSalary']].head())
```

```python
# Import numpy
import numpy as np

# Specify the boundaries of the bins
bins = [-np.inf, 10000, 50000, 100000, 150000, np.inf]

# Bin labels
labels = ['Very low', 'Low', 'Medium', 'High', 'Very high']

# Bin the continuous variable ConvertedSalary using these boundaries
raw_data['boundary_binned'] = pd.cut(raw_data['ConvertedSalary'], bins=bins, labels=labels)

# Print the first 5 rows of the boundary_binned column
print(raw_data[['boundary_binned', 'ConvertedSalary']].head())
```
